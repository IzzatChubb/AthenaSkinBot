```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

from bokeh.io import show, curdoc, output_notebook, push_notebook
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool, Select, Paragraph, TextInput
from bokeh.layouts import widgetbox, column, row
from ipywidgets import interact 

df=pd.read_csv('C:/Users/Izzat/Documents/SEM6/CSP650/Dataset/products_mapped.csv')

#read product types and skin type from other files
temp=pd.read_csv('C:/Users/Izzat/Documents/SEM6/CSP650/Dataset/preprocessed.csv')
temp.label=temp.label.astype(str).str.lower()

option_1=temp.label.unique().tolist()
option_2 = temp.columns[6:].tolist()


output_notebook()
source = ColumnDataSource(df)
plot = figure(x_axis_label = 'T-SNE 1', y_axis_label = 'T-SNE 2', 
              width = 500, height = 400)

plot.circle(x = 'X', y = 'Y', source = source, 
            size = 12, color = '#008080', alpha = .8)

plot.background_fill_color = "beige"
plot.background_fill_alpha = 0.2

hover = HoverTool(tooltips = [
        ('Name', '@name'),
        ('Brand', '@brand'),
        ('Rating','@rank'),
        ('Price(USD) ','@price')],)
plot.add_tools(hover)


def update(op1 = option_1[0], op2 = option_2[0]):
    a_b = op1 + '_' + op2
    new_data = {
        'X' : df[df['label'] == a_b]['X'],
        'Y' : df[df['label'] == a_b]['Y'],
        'label' : df[df['label'] == a_b]['label'],
        'name' : df[df['label'] == a_b]['name'],
        'brand' : df[df['label'] == a_b]['brand'],
        'price' : df[df['label'] == a_b]['price'],
        'rank' : df[df['label'] == a_b]['rank'],
    }
    source.data = new_data
    push_notebook()

    
output_notebook()

interact(update, op1 = option_1, op2 = option_2)
show(plot, notebook_handle = True)
```
